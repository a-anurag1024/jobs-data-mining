```python
# write a program that does k means clustering using sklearn

import numpy as np
from sklearn.cluster import KMeans
from pathlib import Path

# generate some data
X = np.load(Path('C:\\Users\\aanur\\projects\\jobs-data-mining\\mount\\top_skills_embeddings\\embeddings.npy'))

# fit the model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
kmeans.score(X)
```

```python
from tqdm import tqdm
from sklearn.metrics import silhouette_score

scores = []
ks = []
for k in tqdm(range(5, 200, 45)):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    scores.append(silhouette_score(X, kmeans.labels_, metric='cosine')) 
    ks.append(k)
```

```python
import plotly.express as px

fig = px.line(x=ks, y=scores)
fig
```

```python
import json 

with open('C:\\Users\\aanur\\projects\\jobs-data-mining\\mount\\top_skills_embeddings\\metadata.json', 'r') as f:
    metadata = json.load(f)

metadata_ids = {v['array_index']: v for v in metadata}

kmeans = KMeans(n_clusters=200)
kmeans.fit(X)

clusters = {}
for i, l in enumerate(kmeans.labels_):
    metadata_ids[i]['cluster'] = l
    if l not in clusters:
        clusters[l] = [metadata_ids[i]]
    else:
        clusters[l].append(metadata_ids[i])
```

```python
# code for PCA in sklearn 
from sklearn.decomposition import PCA

pca = PCA(n_components=25)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=38)
kmeans.fit(X_pca)

clusters = {}
for i, l in enumerate(kmeans.labels_):
    metadata_ids[i]['cluster'] = l
    if l not in clusters:
        clusters[l] = [metadata_ids[i]]
    else:
        clusters[l].append(metadata_ids[i])
```

```python
silhouette_score(X_pca, kmeans.labels_, metric='cosine')
```

```python
# search for best number of clusters
scores = []
ks = []
for k in tqdm(range(29, 42, 3)):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_pca)
    scores.append(silhouette_score(X_pca, kmeans.labels_, metric='cosine')) 
    ks.append(k)
    
fig = px.line(x=ks, y=scores)
fig
```
